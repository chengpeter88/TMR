```python
import numpy as np
data = np.linspace(-30, 30, 100)
```

```python
data
```

```python
# bonus補充題目

import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

# 這就像是「畫布」的概念
fig = go.Figure()

# go.Scatter：主要的繪圖程式
# add_trace：是plotly一層層疊加上去的合成要素，舉例：
# - add_trace即是主要針對繪圖的繪圖function
# - update_layout即是針對外觀（layout）X、Y軸等要素相關的繪圖function

# 這個的好處：可以輕易的疊加不同的線上去畫布裡面
# 商業意涵：繪製不同模型的利潤折線圖
fig.add_trace(go.Scatter(x=data, y=data**2,
                mode='lines'))

fig.add_trace(go.Scatter(x=data, y=data**3,
                mode='lines'))

iplot(fig)
```

```python
# 如果要加入X、Y軸的要素，可以使用update_layout進行疊加

fig.update_layout(
    title="Parabola",
    xaxis_title="X",
    yaxis_title="Y",
    )

iplot(fig)

# Go與px方法主要的差異在，當我們要進行多線繪圖的時候，go的方式可以更容易且直觀的進行多線繪圖
```
